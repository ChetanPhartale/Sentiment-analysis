**Data Cleaning or Pre-Processing**

**Lower Case**

```python
data.tweet = data.tweet.str.lower()
```

**Remove URL**

```python
data.tweet = data.tweet.str.replace(r'http\S+|www.\S+','',case=False)
data.iloc[991,0]
```

**Remove Punctuations**

```python
import string
string.punctuation
```

```python
def remove_punctuation(abc):
    text_nopunt="".join([c  
                         for c in abc  
                         if c not in string.punctuation])
    return text_nopunt
```

```python
data.iloc[991,0]
```

```python
data.tweet=data.tweet.apply(lambda x : remove_punctuation(x))  
```

```python
data.iloc[991,0]
```

**Remove Number**

```python
data['tweet'] = data['tweet'].str.replace('\d+','')  
```

**Removing Stop Words**

```python
# Import stopwords with nltk.
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop
```

```python
data['tweet'] = data['tweet'].apply(lambda x: ' '.join([abc   
                                                      for abc in x.split() 
                                                      if abc not in (stop)]))
```

**Remove Common words**

```python
import re 
nltk.download('words')   
words = set(nltk.corpus.words.words()) 
words
```

```python
from nltk.stem import WordNetLemmatizer 
import nltk
nltk.download('wordnet') 

lemmatizer = WordNetLemmatizer() 
word_tokenizer = nltk.tokenize.WhitespaceTokenizer() 
```

```python
def lemmatize_text(tweet):
    return [lemmatizer.lemmatize(w,"v") 
            for w in word_tokenizer.tokenize(tweet)]
```

**Strip Extra White Space**

```python
data['tweet'] = (data['tweet'].astype("str").str.rstrip())
data.iloc[2000,0]
```

**Finished Data Cleaning**

**Term Document Matrix**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english') 
data_cv = cv.fit_transform(data.tweet)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names()) 

tdm = data_dtm.transpose()
tdm.head()
```

```python
tdm['freq'] = tdm.sum(axis=1)
tdm.head()
```

```python
tdm.reset_index(inplace=True)
tdm.head()
```

```python
tdm1=tdm[["index","freq"]] 
tdm1.rename(columns = {'index':'Word'}, inplace = True) 
tdm1.sort_values(by='freq',ascending=False,inplace=True)  
tdm1.head(20)
```

**WordCloud**

```python
text = " ".join(review for review in data.text)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

stopwords = set(STOPWORDS)

cloud = WordCloud(stopwords=stopwords, 
                      background_color="white",
                      collocations=False,
                      max_words=100).generate(text)

plt.imshow(cloud, interpolation='bilinear')  
plt.axis("off") 
plt.show() 
```

**Sentiment Analysis**

```python
data
```

```python
data = data.drop_duplicates('tweet') 
data
```

```python
from textblob import TextBlob

pol = lambda x: TextBlob(x).sentiment.polarity 
data['polarity'] = data['tweet'].apply(pol)
data.head(5)
```

```python
data['Sentiment'] = np.where(data['polarity']>= 0, 'Positive', 'Negative')
data.head()
```

```python
data.Sentiment.value_counts().plot.bar()
```
